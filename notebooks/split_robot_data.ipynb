```python
df = pd.read_csv('/home/ss26/Projects/Smart-Tools/data/S2023_Yaskawa_Processed.csv')
df
```

```python
df0 = df.loc[df['Activity'] == 0]
df1 = df.loc[df['Activity'] == 1]
df2 = df.loc[df['Activity'] == 2]
df3 = df.loc[df['Activity'] == 3]
```

```python
train_num = lambda x: int(0.7*x)
```

```python
def make_train_test_val_split(df0, df1, df2, df3):
    df_train = pd.DataFrame()
    df_val = pd.DataFrame()
    df_test = pd.DataFrame()

    # engraving
    tv_idx = train_num(df0.shape[0])
    df_train = pd.concat(
        [df_train, df0.iloc[:tv_idx//2, :]], ignore_index=True)
    df_val = pd.concat(
        [df_val, df0.iloc[tv_idx//2:tv_idx, :]], ignore_index=True)
    df_test = pd.concat([df_test, df0.iloc[tv_idx:, :]], ignore_index=True)

    # cutting
    tv_idx = train_num(df1.shape[0])
    df_train = pd.concat(
        [df_train, df1.iloc[:tv_idx//2, :]], ignore_index=True)
    df_val = pd.concat(
        [df_val, df1.iloc[tv_idx//2:tv_idx, :]], ignore_index=True)
    df_test = pd.concat([df_test, df1.iloc[tv_idx:, :]], ignore_index=True)

    # sanding
    tv_idx = train_num(df2.shape[0])
    df_train = pd.concat(
        [df_train, df2.iloc[:tv_idx//2, :]], ignore_index=True)
    df_val = pd.concat(
        [df_val, df2.iloc[tv_idx//2:tv_idx, :]], ignore_index=True)

    df_test = pd.concat([df_test, df2.iloc[tv_idx:, :]], ignore_index=True)

    # routing
    tv_idx = train_num(df3.shape[0])
    df_train = pd.concat([df_train, df3.iloc[:tv_idx//2, :]], ignore_index=True)
    df_val = pd.concat([df_val, df3.iloc[tv_idx//2:tv_idx, :]], ignore_index=True)
    df_test = pd.concat([df_test, df3.iloc[tv_idx:, :]], ignore_index=True)

    return df_train, df_val, df_test


df_train, df_val, df_test = make_train_test_val_split(df0, df1, df2, df3)
```

```python
import sktime as st
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
    import pyarrow as pa
```

```python
# pf = pa.parquet.ParquetFile('/home/ss26/Projects/Smart-Tools/data/F2021_processed.parquet')

data = pa.dataset.dataset('/home/ss26/Projects/Smart-Tools/data/F2021_processed.parquet').scanner().head(1000).to_pandas()
print(data.shape)
data
```

```python
start_date_string = "14/10/2021 08:00:00.000"   
end_date_string = "14/10/2021 18:00:00.000"     

# Convert the strings of dates into the Python datetime format
start_datetime = datetime.strptime(start_date_string, '%d/%m/%Y %H:%M:%S.%f')
end_datetime = datetime.strptime(end_date_string, '%d/%m/%Y %H:%M:%S.%f')

# Create a range of dates
time_index = pd.date_range(start = start_datetime, end = end_datetime, freq="L")
```

```python
time_index = time_index[:data.shape[0]]
range_index = pd.RangeIndex(0,1000)
print(f"Time index shape: {time_index.shape}")
print(f"Time index type: {type(time_index)}")

print(f"Range index shape: {range_index.shape}")
print(f"Range index type: {type(range_index)}")
```

```python
data["Index"] = range_index
data["Timestamp"] = time_index
data = data.set_index(["Index", "Timestamp"])
data
```

```python
data = data.sort_values(by='Activity')

data0 = data.loc[data["Activity"] == 0]
data1 = data.loc[data["Activity"] == 1]
data2 = data.loc[data["Activity"] == 2]
data3 = data.loc[data["Activity"] == 3]
```
